*  **4가지 분야의 기술을 사용하는 4가지 제품 및 서비스**
 1. 언어
 2. 음성
 3. 이미지
 4. 자율주행
-------------
1. 언어
 + 네이버 파파고   
네이버가 자체적으로 개발한 인공신경망 기반의 기계 번역 기술인 NMT를 이용한 번역 서비스. 입력된 문자를 쪼개어 번역하는 것이 아니라 문구를 통째로 번역하기 때문에, 맥락이 고려된 자연스러운 번역 결과를 얻을 수 있습니다.    

 + 셀비 챗봇    
 인공지능 기술 기반의 대화형 상담 에이전트 시스템으로 자연어 처리 기술을 통한 대화 관리가 가능함. 채팅 인터페이스를 통해 사용자와 상호작용하는 지능형 대화 플랫폼으로, 주요 SNS 메신저 제품에 적용하여 사용할 수 있습니다.

2. 음성
 + 바이두 딥 보이스   
 인공지능과 딥 러닝 기술을 적용한 음성 모방 솔루션, 3.7초의 음성 데이터만 있으면 음성을 복제하는데 가능하며 별도의 녹음 없이 자동화하여 저자의 목소리로 책을 읽어주는 도구를 개발했습니다.

  + 아마존 알락사와 아마존 에코   
  아마존 에코를 이용해 알락사와 의사소통을 할 수 있으며 음악재생, 알람설정, 날씨정보 제공 등 많은 기능들을 제공해줍니다.

3. 이미지
 + 인도네시아 스타트업 그린고   
 쓰레기 문제 해결을 위해 컴퓨터 비전 기술을 통해 쓰레기 종류를 파악하고 처리 시세 데이터와 연결해 주어 재활용 쓰레기 가치를 모르는 수거 업자들에게 재활용 쓰레기 가치를 알려줍니다.

  + Nvidia Gaugan   
  장면의 레이아웃을 묘사하는 레이블이 지정된 스케치인 분할 지도로부터 포토리얼리스틱 이미지를 제작하는 어플리케이션이다. 강, 풀, 바위 등 지정된 색으로 색을 칠하면 그림을 그려줍니다.


4. 자율주행
 + 테슬라 오토파일럿   
 8개의 서라운드 카메라로 최대 360도 시야를 확보하고 12개의 초음파 센서로 시야를 보완함. 내비게이션과 연동 되어 경로를 최적화 하고 저속 차량 또는 트럭 뒤에서 주행하지 않도록 차선 변경을 제안함. 스마트 호출로 주차장 내에서 운전자가 위치한 곳으로 차량을 호출 할 수 있습니다.
